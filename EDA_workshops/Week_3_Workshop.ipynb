```python
# Isolate numerical columns
numbersdf = df.select_dtypes(include=np.number)

#Perform outlier detection on numerical columns using scipy
numbersdf[(np.abs(stats.zscore(numbersdf)) < 3).all(axis=1)]

#Recombine cleaned numbersdf with df
df = df[df.index.isin(numbersdf.index)]
```

```python
#Summary stats
df.describe()
```

```python
# Calculate IQR for numeric columns only
iqr_results = df.select_dtypes(include=np.number).apply(lambda x: np.quantile(x, 0.75) - np.quantile(x, 0.25))

print(iqr_results)
```

```python
# One-hot encode the 'category' column
df_encoded = pd.get_dummies(df['Location '])

# Combine the original DataFrame with the encoded columns, removing the original location column
df_final = pd.concat([df.drop('Location ', axis= 1), df_encoded], axis=1)

print(df_final)
```

```python
#Define a correlation matrix
corr_matrix = df_final.corr()

print("Correlation matrix:", corr_matrix)
```

Upon visual inspection, no column seems redundant

```python
#Create a heatmap based on the correlation matrix
sns.set(rc={'figure.figsize': (20, 8)}) #Set larger figure size so values are readable
sns.heatmap(corr_matrix, annot=True, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
```

S-N and S-Ash seem to have strong positive correlations.
Sulfur, nitrogen, and ash are all mainly inorganic compounds so it makes sense for them to be correlated.

VM-Ash and C-O seem to have strong negative correlations.
This makes sense becuase VM is mostly organic compounds and Ash is mostly inorganic. The same hold true in that C is organic and O is inorganic.

Plant capacity has a notable positive correlation with minimum selling price. This makes sense in the context of economies of scale.

Chinese production seems to have a negative correlation with minimum selling price.

```python
# Visualize the relationships between correlated numerical variables using a scatter plot
plt.scatter(df_final['S (%)'], df_final['Ash (%)'])
plt.xlabel("S (%)")
plt.ylabel("Ash (%)")
plt.title("Sulfur vs. Ash")
plt.show()

plt.scatter(df_final['VM (%)'], df_final['Ash (%)'])
plt.xlabel("VM (%)")
plt.ylabel("Ash (%)")
plt.title("Volatile Matter vs. Ash")
plt.show()

plt.scatter(df_final['C (%)'], df_final['O (%)'])
plt.xlabel("C (%)")
plt.ylabel("O (%)")
plt.title("Carbon vs. Oxygen")
plt.show()

plt.scatter(df_final['S (%)'], df_final['N (%)'])
plt.xlabel("S (%)")
plt.ylabel("N (%)")
plt.title("Sulfur vs. Nitrogen")
plt.show()

plt.scatter(df_final['Plant capacity (kg/hr)'], df_final['MSP'])
plt.xlabel("Plant capacity (kg/hr)")
plt.ylabel("MSP")
plt.title("Plant Capacity vs. Minimum Selling Price")
plt.show()
```
